```python



import cv2
from matplotlib import pyplot as plt
```

```python
img = 'img/puppywithball.jpg'
def img_show(imagename,title):
  image = cv2.imread(imagename)
  im_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  plt.imshow(im_rgb)
  plt.title(title)
  plt.grid(False)
  plt.show()


```

```python
img_show(img,'Image')
```
