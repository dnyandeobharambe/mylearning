```python



import cv2
from matplotlib import pyplot as plt
```

```python
img = 'img/puppywithball.jpg'
image = cv2.imread(img)
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
# get height and width
(h, w) = image.shape[:2]
plt.imshow(image)
plt.title('original')
plt.grid(False)
plt.show()

# get center 
(X,Y) = (w//2,h//2)

testimg1 = image[0:Y,0:X]
plt.imshow(testimg1)
plt.title('top-left-corner')
plt.grid(False)
plt.show()

testimg2 = image[0:Y,X:w]
plt.imshow(testimg2)
plt.title('top-right-corner')
plt.grid(False)
plt.show()


testimg3 = image[Y:h,X:w]
plt.imshow(testimg3)
plt.title('bottom-right-corner')
plt.grid(False)
plt.show()

testimg4 = image[Y:h,0:X]
plt.imshow(testimg4)
plt.title('bottom-left-corner')
plt.grid(False)
plt.show()

# this is R G B
image[Y:h,0:X] = (255,0,0)
plt.imshow(image)
plt.title('updated-image')
plt.grid(False)
plt.show()

  

```
